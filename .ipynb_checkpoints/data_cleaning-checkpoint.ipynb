# [Kaggle Project: House Prices](https://www.kaggle.com/competitions/house-prices-advanced-regression-techniques)


```python
import pandas as pd
import numpy as np
```

```python
df_train = pd.read_csv('data/train.csv')
df_test  = pd.read_csv('data/test.csv')
```

## Data Cleaning

```python
display(df_train[:5])
print('The shape of df_train is ',df_train.shape)
display(df_test[:5])
print('The shape of df_test is ',df_test.shape)
```

### Drop Id

```python
df_train.drop(columns=['Id'],inplace=True)
df_test.drop(columns=['Id'],inplace=True)
```

### Missing Values

```python
cols = df_test.columns
```

```python
df_null = pd.DataFrame({
    'train_null%':df_train[cols].isna().sum().values/df_train[cols].shape[0]*100,
    'test_null%':df_test[cols].isna().sum().values/df_test[cols].shape[0]*100},
    index = cols)

row_idx = (df_null['train_null%']>0) | (df_null['test_null%']>0)
col_idx = df_null.columns

df_null_sub = df_null.loc[row_idx,col_idx]
df_null_sub['sum%'] = df_null_sub.sum(axis=1)
```

```python
df_null_sub.sort_values(by='sum%')
```

```python
drop_cols = list(df_null_sub.loc[df_null_sub['sum%']>1].index)
# drop_cols
```

```python
df_train.drop(columns=drop_cols,inplace=True)
df_test.drop(columns=drop_cols,inplace=True)
print('The shape of df_train is ',df_train.shape)
print('The shape of df_test is ',df_test.shape)
```

### Fill Missing Values

```python
train_cols_fillna = df_null_sub.loc[df_null_sub['sum%']<=1].loc[df_null_sub['train_null%']>0].index
test_cols_fillna = df_null_sub.loc[df_null_sub['sum%']<=1].loc[df_null_sub['test_null%']>0].index
```

#### train data

```python
df_train[train_cols_fillna].dtypes
```

```python
df_train[train_cols_fillna].value_counts()/df_train.shape[0]
```

```python
df_train.fillna(
    value={train_cols_fillna[0]:df_train[train_cols_fillna].value_counts().index[0][0]},
    inplace=True)
```

#### test data

```python
test_cols_fillna_numeric = df_test[test_cols_fillna].select_dtypes([np.number]).columns
test_cols_fillna_object = df_test[test_cols_fillna].select_dtypes([object]).columns
```

```python
df_test[test_cols_fillna_numeric].describe()
```

```python
df_test[test_cols_fillna_numeric] = df_test[test_cols_fillna_numeric].fillna(df_test[test_cols_fillna_numeric].mean())
```

```python
df_test[test_cols_fillna_object].describe()  
```

```python
from sklearn.impute import SimpleImputer
```

```python
imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')

test_dtypes = df_test.dtypes

df_test = pd.DataFrame(imp.fit_transform(df_test), 
                        index= df_test.index, 
                        columns = df_test.columns).astype(test_dtypes.to_dict())
```

```python
df_train.to_csv('df_train.csv', index = False, header=True)
df_test.to_csv('df_test.csv', index = False, header=True)
```
